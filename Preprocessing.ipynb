```python
#contains user actions in prior set (add_to_cart_order and reordered)
#reordered: this product is bought again by the same customers

zf = ZipFile('order_products__prior.csv.zip') 
df_order_products_prior = pd.read_csv(zf.open('order_products__prior.csv'))
```

```python
df_order_products_prior.head()
```

```python
#contains user actions in train set (add_to_cart_order and reordered)
zf = ZipFile('order_products__train.csv.zip') 
df_order_products_train = pd.read_csv(zf.open('order_products__train.csv'))
```

```python
df_order_products_train.head()
```

```python
zf = zipfile.ZipFile('orders.csv.zip') 
df_orders = pd.read_csv(zf.open('orders.csv'))
```

```python
#fill all NaN in days_since_prior_order with 0 
#and check if the missing data is being replaced
df_orders = df_orders.fillna(0)
df_orders.isna().sum()
```

```python
#the statistical behavior doesnt change too much before filling zeroes in it
#should be a possible way to processing it
df_orders.describe()
```

```python
df_prior_data = df_orders[df_orders['eval_set']=='prior'].merge(df_order_products_prior,how='left',on='order_id')
```

```python
df_prior_data = df_prior_data.merge(df_products[["product_id","aisle_id","department_id"]],how='left',on='product_id')
```

```python
df_prior_data
```

```python
df_train_data = df_orders[df_orders['eval_set']=='train'].merge(df_order_products_train,how='left',on='order_id')
```

```python
df_train_data = df_train_data.merge(df_products[["product_id","aisle_id","department_id"]],how='left',on='product_id')
```

```python
df_train_data
```

```python
df_prior_data = df_prior_data.drop(["order_id","eval_set"],axis = 1)
df_train_data = df_train_data.drop(["order_id","eval_set"],axis = 1)
```

```python
df_prior_data
```

```python
move_to_end = ['add_to_cart_order','reordered']
```

```python
df_prior_data = df_prior_data[[c for c in df_prior_data if c not in move_to_end] + move_to_end]
```

```python
df_prior_data.head()
```

```python
df_train_data = df_train_data[[c for c in df_train_data if c not in move_to_end] + move_to_end]
```

```python
df_train_data.head()
```

```python
df_test_data = df_orders[df_orders['eval_set']=='test']
```

```python
df_test_data.head(20)
```

```python
filename = "train_dataset"
compression_options = dict(method='zip', archive_name=f'{filename}.csv')
df_prior_data.to_csv(f'{filename}.zip', compression=compression_options)
```

```python
filename = 'validation_dataset'
compression_options = dict(method='zip', archive_name=f'{filename}.csv')
df_train_data.to_csv(f'{filename}.zip', compression=compression_options)
```

```python
filename = 'test_dataset'
compression_options = dict(method='zip', archive_name=f'{filename}.csv')
df_test_data.to_csv(f'{filename}.zip', compression=compression_options)
```
